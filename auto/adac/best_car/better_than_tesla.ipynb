```python
from find_best_car import (
    Column, 
    get_scored_df,
    _get_fixed_column_name,
)


df = get_scored_df(de_discount=False, feature_file_name="feature.csv")
```

```python
col_fixed_adac = _get_fixed_column_name(Column.ADAC_PAKET)

teslas = df[df[Column.NAME].str.contains('Tesla')]
teslas_with_adac = teslas[teslas[col_fixed_adac].notna()]
MAX_ADAC_ADDITIONAL_PRICE = max(teslas_with_adac[col_fixed_adac])
MAX_ADAC_ADDITIONAL_PRICE
```

# Cars where ADAC packet price is less then Tesla's

```python
tesla = teslas.iloc[0]

parking_assists = {
            "vo.+hi. mit Front- und Heckkamera",
            "Front- und Heckkamera",
            "vo.+hi. mit Rückfahrkamera",
            "hinten mit Rückfahrkamera",
}

df_best = df[
    (df[col_fixed_adac] <= MAX_ADAC_ADDITIONAL_PRICE) & \
    (df[Column.TRANSMISSION] != "Schaltgetriebe") & \
    (df[_get_fixed_column_name(Column.ACCELERATION)] <= 9.8) & \
    (df["Autom. Abstandsregelung"] == tesla["Autom. Abstandsregelung"]) & \
    (df["Regensensor"] == tesla["Regensensor"]) & \
    (df["Lichtsensor"] == tesla["Lichtsensor"]) & \
    (df[_get_fixed_column_name("Kofferraumvolumen normal")] >= 380) & \
    (df["Einparkhilfe - Bezeichnung"].isin(parking_assists)) & \
    (df[Column.TOTAL_SCORE] >= tesla[Column.TOTAL_SCORE])
].reset_index(drop=True)
df_best.to_excel("/tmp/better_tesla.xlsx")
df_best[[Column.ID, Column.NAME, Column.TOTAL_SCORE, Column.EURO_PER_SCORE]]
```
