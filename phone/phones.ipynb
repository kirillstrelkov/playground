```python
from find_best_phone import get_filtered_df, get_phone_df, get_weited_df
import seaborn as sns

def coloraze(df):
    cm = sns.light_palette("seagreen", as_cmap=True)
    return df.style.background_gradient(cmap=cm, axis=1)
```

```python
coloraze(get_filtered_df(get_phone_df()).drop(['weight', 'reversed'], axis=1))
```
