```python
from find_best_car import (
    Column, 
    get_scored_df,
    _get_fixed_column_name,
)

df = get_scored_df(feature_file_name="feature_parents.csv")

```

# Best cars by score

```python
FIXED_PRICE_COLUMN = _get_fixed_column_name(Column.PRICE)
VIEW_COLUMNS = [Column.NAME, FIXED_PRICE_COLUMN, Column.TOTAL_SCORE, Column.EURO_PER_SCORE]
VIEW_LIMIT = 40

df[VIEW_COLUMNS].head(VIEW_LIMIT)
```

# Best cars by euro per score

```python
df[VIEW_COLUMNS].sort_values([Column.EURO_PER_SCORE, Column.TOTAL_SCORE]).head(VIEW_LIMIT)
```

```python
best_grandland = df[df[Column.NAME].str.contains('Grandland')].iloc[0][VIEW_COLUMNS]
grandland_score = best_grandland[Column.TOTAL_SCORE]
grandland_euro_per_score = best_grandland[Column.EURO_PER_SCORE]
best_grandland
```

# Better than Grandland by points

```python
d = df[df[Column.TOTAL_SCORE] >= grandland_score]
d[VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```

# Better than Grandland by points and euro per point

```python
d = df[(df[Column.TOTAL_SCORE] >= grandland_score) & (df[Column.EURO_PER_SCORE] <= grandland_euro_per_score)]
d[VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```

# Better than Grandland by points and euro per point and cheaper

```python
d = df[
    (df[Column.TOTAL_SCORE] >= grandland_score) & 
    (df[Column.EURO_PER_SCORE] <= grandland_euro_per_score) & 
    (df[FIXED_PRICE_COLUMN] <= best_grandland[FIXED_PRICE_COLUMN])
]
d[VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```
