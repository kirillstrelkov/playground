# Most suitable car between all cars

```python
from find_best_car import (
    Column, 
    get_scored_df,
    _get_fixed_column_name,
)


df = get_scored_df(de_discount=False, feature_file_name="feature.csv")
```

```python
my_cars = df[df[Column.NAME].str.contains('Subaru Impreza')]

top_car = my_cars.iloc[0]
top_car_euro_per_score = top_car[Column.EURO_PER_SCORE]
top_car_total_score = top_car[Column.TOTAL_SCORE]

# imprezas
top_car
```

### Better in score

```python
FIXED_PRICE_COLUMN = _get_fixed_column_name(Column.PRICE)
VIEW_COLUMNS = [Column.NAME, FIXED_PRICE_COLUMN, Column.TOTAL_SCORE, Column.EURO_PER_SCORE]
VIEW_LIMIT = 10

df[df[Column.TOTAL_SCORE] > top_car_total_score][VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```

### Better in euro per score

```python
df[df[Column.EURO_PER_SCORE] < top_car_euro_per_score][VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```

### Better in everything

```python
d = df[(df[Column.TOTAL_SCORE] >= top_car_total_score) & (df[Column.EURO_PER_SCORE] <= top_car_euro_per_score)]
d[VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```

### Better in everything and cheaper

```python
d = df[
    (df[Column.TOTAL_SCORE] >= top_car_total_score) & 
    (df[Column.EURO_PER_SCORE] <= top_car_euro_per_score) &
    (df[FIXED_PRICE_COLUMN] <= top_car[FIXED_PRICE_COLUMN])
]
d[VIEW_COLUMNS].reset_index(drop=True).head(VIEW_LIMIT)
```
